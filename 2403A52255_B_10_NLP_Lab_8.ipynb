**Corpus**

```python
D1="I am working as a professor"
D2="I am working in SR University"
D3="I did my phd in NITW"
D4="I did my masters in OUCE"
D5="SR University is located in Warangal"
D6="NITW is a prestigious institute"
D7="OUCE offers various engineering courses"
D8="I completed my graduation in Hyderabad"
D9="Warangal is a historic city"
D10="Hyderabad is the capital of Telangana"
```

**Combine Corpus**

```python
documents = [D1,D2,D3,D4,D5,D6,D7,D8,D9,D10]

combined_text = " ".join(documents)
words = combined_text.lower().split()
```

**Unigram** **Counts**

```python
import collections

unigram_counts = collections.Counter(words)

print("Unigram Counts:")
for word, count in unigram_counts.most_common():
    print(word, ":", count)

V = len(unigram_counts)
print("Vocabulary Size =", V)
```

**Bigram** **Counts**

```python
bigrams = []
for i in range(len(words)-1):
    bigrams.append((words[i], words[i+1]))

bigram_counts = collections.Counter(bigrams)

print("\nBigram Counts:")
for bg, count in bigram_counts.most_common():
    print(bg[0], bg[1], ":", count)
```

**Trigram Counts**

```python
trigrams = []
for i in range(len(words)-2):
    trigrams.append((words[i], words[i+1], words[i+2]))

trigram_counts = collections.Counter(trigrams)

print("\nTrigram Counts:")
for tg, count in trigram_counts.most_common():
    print(tg[0], tg[1], tg[2], ":", count)
```

**Bigram Next Word Prediction**

```python
def predict_next_word_bigram(sentence):

    words_seq = sentence.lower().split()
    last_word = words_seq[-1]

    candidates = {}
    for (w1,w2),count in bigram_counts.items():
        if w1 == last_word:
            candidates[w2] = count

    if not candidates:
        return "No prediction available"

    predicted = max(candidates, key=candidates.get)
    return predicted
```

**Deployment of Bi-Gram Model**

```python
print(predict_next_word_bigram("I am"))
print(predict_next_word_bigram("Warangal is"))
print(predict_next_word_bigram("Hyderabad is"))
```

**Trigram Next Word Prediction**

```python
def predict_next_word_trigram(sentence):

    words_seq = sentence.lower().split()
    if len(words_seq) < 2:
        return "Enter at least two words"

    last_two = tuple(words_seq[-2:])
    candidates = {}

    for (w1,w2,w3),count in trigram_counts.items():
        if (w1,w2) == last_two:
            candidates[w3] = count

    if not candidates:
        return "No prediction available"

    predicted = max(candidates, key=candidates.get)
    return predicted
```

**Deployement** **of Tri Gram Model**

```python
print(predict_next_word_trigram("I am working"))
print(predict_next_word_trigram("Warangal is a"))
print(predict_next_word_trigram("Hyderabad is the"))
```

**Next Word Prediction Using Bi-Gram Counts with Laplace Smoothening**

```python
def predict_bigram_laplace(sentence):

    words_seq = sentence.lower().split()
    last_word = words_seq[-1]

    max_prob = 0
    best_word = None

    for word in unigram_counts:
        bigram_count = bigram_counts.get((last_word,word),0)
        prob = (bigram_count + 1) / (unigram_counts[last_word] + V)

        if prob > max_prob:
            max_prob = prob
            best_word = word

    return best_word
```

**Deployment**

```python
print("Input: I am")
print("Predicted:", predict_bigram_laplace("I am"))

print("\nInput: I did")
print("Predicted:", predict_bigram_laplace("I did"))

print("\nInput: Warangal is")
print("Predicted:", predict_bigram_laplace("Warangal is"))

print("\nInput: Hyderabad is")
print("Predicted:", predict_bigram_laplace("Hyderabad is"))

print("\nInput: SR University")
print("Predicted:", predict_bigram_laplace("SR University"))
```

**Next Word Prediction Using Tri-Gram Counts based on laplace smoothening**

```python
def predict_trigram_laplace(sentence, trigram_counts, bigram_counts, unigram_counts, V):

    words_seq = sentence.lower().split()

    # Need at least 2 words
    if len(words_seq) < 2:
        return "Enter at least two words"

    last_two = tuple(words_seq[-2:])

    # If bigram not found
    if last_two not in bigram_counts:
        return "Context not found in corpus"

    max_prob = 0
    best_word = None

    for word in unigram_counts:   # iterate over full vocabulary

        trigram_count = trigram_counts.get((last_two[0], last_two[1], word), 0)
        bigram_count = bigram_counts.get(last_two, 0)

        # Laplace Formula:
        # P(w3 | w1,w2) = (Count(w1,w2,w3) + 1) / (Count(w1,w2) + V)

        prob = (trigram_count + 1) / (bigram_count + V)

        if prob > max_prob:
            max_prob = prob
            best_word = word

    return best_word
```

**Deployment**

```python
print("Input: I am working")
print("Predicted:", predict_trigram_laplace(
    "I am working",
    trigram_counts,
    bigram_counts,
    unigram_counts,
    V))

print("\nInput: I did my")
print("Predicted:", predict_trigram_laplace(
    "I did my",
    trigram_counts,
    bigram_counts,
    unigram_counts,
    V))

print("\nInput: Warangal is a")
print("Predicted:", predict_trigram_laplace(
    "Warangal is a",
    trigram_counts,
    bigram_counts,
    unigram_counts,
    V))
```

**Next Word Prediction Using Bi-Gram Counts with Add - K Smoothening**

```python
def predict_bigram_addk(sentence, bigram_counts, unigram_counts, V, K=0.5):

    words_seq = sentence.lower().split()

    if not words_seq:
        return "Enter a valid sentence"

    last_word = words_seq[-1]

    if last_word not in unigram_counts:
        return "Word not found in corpus"

    max_prob = 0
    best_word = None

    for word in unigram_counts:

        bigram_count = bigram_counts.get((last_word, word), 0)
        unigram_count = unigram_counts.get(last_word, 0)

        prob = (bigram_count + K) / (unigram_count + K * V)

        if prob > max_prob:
            max_prob = prob
            best_word = word

    return best_word
```

** Delpoyment**

```python
print("Input: I am")
print("Predicted:", predict_bigram_addk(
    "I am",
    bigram_counts,
    unigram_counts,
    V,
    K=0.5))

print("\nInput: I did")
print("Predicted:", predict_bigram_addk(
    "I did",
    bigram_counts,
    unigram_counts,
    V,
    K=0.5))

print("\nInput: Warangal is")
print("Predicted:", predict_bigram_addk(
    "Warangal is",
    bigram_counts,
    unigram_counts,
    V,
    K=0.5))
```

**Next Word Prediction Using Tri-Gram Counts with Add - K Smoothening**

```python
def predict_trigram_addk(sentence, trigram_counts, bigram_counts, unigram_counts, V, K=0.5):

    words_seq = sentence.lower().split()

    # Need at least two words
    if len(words_seq) < 2:
        return "Enter at least two words"

    last_two = tuple(words_seq[-2:])

    # If context not found
    if last_two not in bigram_counts:
        return "Context not found in corpus"

    max_prob = 0
    best_word = None

    # Check probability for every word in vocabulary
    for word in unigram_counts:

        trigram_count = trigram_counts.get((last_two[0], last_two[1], word), 0)
        bigram_count = bigram_counts.get(last_two, 0)

        # Add-K formula
        prob = (trigram_count + K) / (bigram_count + K * V)

        if prob > max_prob:
            max_prob = prob
            best_word = word

    return best_word
```

**Deployment**

```python
print("Input: I am working")
print("Predicted:", predict_trigram_addk(
    "I am working",
    trigram_counts,
    bigram_counts,
    unigram_counts,
    V,
    K=0.5))

print("\nInput: I did my")
print("Predicted:", predict_trigram_addk(
    "I did my",
    trigram_counts,
    bigram_counts,
    unigram_counts,
    V,
    K=0.5))

print("\nInput: Warangal is a")
print("Predicted:", predict_trigram_addk(
    "Warangal is a",
    trigram_counts,
    bigram_counts,
    unigram_counts,
    V,
    K=0.5))
```
